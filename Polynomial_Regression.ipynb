# Generate a Non linear quadratic dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 100

X = np.linspace(-10,10,n_samples)
y = 3*X**2 + 2*X + 5 + np.random.randn(n_samples)*10
X = X.reshape(100, 1)

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Non-linear Quadratic Dataset")
plt.show()
```

# Fit Linear Regression vs Polynomial Regression (using PolynomialFeatures).

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X,y)
y_pred = lr.predict(X)

print(mean_squared_error(y,y_pred))
```

```python
plt.scatter(X,y,label='True data')
plt.plot(X,y_pred,color='red',label='Linear fit')
plt.title("Linear Regression on quadratic dataset (Underfitting)")
plt.show()
```

```python
X.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures(degree=30)
X_poly = poly.fit_transform(X)
lr = LinearRegression()
lr.fit(X_poly, y)
y_pred = lr.predict(X_poly)

print(mean_squared_error(y,y_pred))
```

```python
plt.scatter(X.flatten(), y, label="True Data")
plt.plot(X.flatten(), y_pred, color='red', label="Quadratic Fit")
plt.title("Linear Regression on polynomial features (Overfitting)")
plt.legend()
plt.show()
```
