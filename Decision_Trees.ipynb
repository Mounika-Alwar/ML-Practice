# Train a DT, Visualize the Tree and Show how changing max_depth affects bias Vs Variance

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def train_and_plot(max_depth):
    clf = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    clf.fit(X_train, y_train)

    y_train_pred = clf.predict(X_train)
    y_test_pred = clf.predict(X_test)

    print(f"\n---- max_depth = {max_depth} ----")
    print("Train Accuracy:", accuracy_score(y_train, y_train_pred))
    print("Test Accuracy :", accuracy_score(y_test, y_test_pred))

    plt.figure(figsize=(10, 6))
    plot_tree(clf,
              feature_names=iris.feature_names,
              class_names=iris.target_names,
              filled=True,
              rounded=True,
              fontsize=9)
    plt.title(f"Decision Tree (max_depth={max_depth})")
    plt.show()
```

```python
for depth in [1, 2, 3, 4, None]:
    train_and_plot(depth)
```
